```python
def loadFromKaggle():

        bakery_data = pd.read_csv('BreadBasket_DMS.csv', encoding='utf-8')
        bakery_data['Date Time'] = bakery_data['Date'] + " " + bakery_data['Time']
        bakery_data = bakery_data.drop(['Date', 'Time'], axis=1)
        bakery_data = bakery_data.drop(['Date Time'], axis=1)
        bakery_data = bakery_data[~bakery_data['Item'].str.contains('NONE')]

        tdl = []
        for i in range(1, bakery_data.Transaction.count() + 1):
            tdf = bakery_data[bakery_data.Transaction == i]
            l = set()
            for j in range(0, tdf.Transaction.count()):
                l.add(tdf.Item.iloc[j])
            if len(l) > 0:
                tdl.append(list(l))
            else:
                tdl.append(None)

        col = ['items']
        TDB = pd.DataFrame({"items": tdl}, columns=col)
        TDB = TDB.dropna()
        return TDB
    
def loadFromIbm():
    # read data from Ibm
    with open('data', encoding='utf-8') as f:
        content = f.readlines()

    content = [x.strip() for x in content]

    Transaction = []  # to store transaction
    Frequent_items_value = {}  # to store all frequent item sets

    # to fill values in transaction from txt file
    Transaction = {}
    Transaction
    for i in range(0, len(content)):
        rowd = content[i].split(' ')
        rowd = [r for r in rowd if r != '']
        rowd = rowd[1:]
        if Transaction.get(rowd[0], None) == None:
            Transaction[rowd[0]] = [rowd[1]]
        else:
            Transaction[rowd[0]].append(rowd[1])

    col = ['items']
    TDB = pd.DataFrame({"items": list(Transaction.values())}, columns=col)
    TDB = TDB.dropna()
    return TDB

def make_weka(TDB,fn):
    maxItemIndex = len(max(TDB['items'].tolist(), key=len))
    weka_data = pd.DataFrame(columns=["Item_" + str(i+1) for i in range(maxItemIndex)])
    itemdata = []
    for itemset in TDB['items']:
        itemList = []
        [itemList.insert(index,item) for index,item in enumerate(itemset)]        
        itemdata.append(itemList)       
        [itemList.append(0) for i in range(maxItemIndex - len(itemList))]
        
    weka_data = pd.DataFrame(itemdata,columns=["Item_" + str(i+1) for i in range(maxItemIndex)])
    weka_data.to_csv(fn,index=0)
        
TDB = loadFromKaggle()
make_weka(TDB,'Weka_BreadBasket_DMS.csv')
    
TDB = loadFromIbm()
make_weka(TDB,"Weka_data.csv")
```
